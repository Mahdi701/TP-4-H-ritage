```python
import numpy as np
import matplotlib.pyplot as plt

class Signal:
    def __init__(self, nbr):
        self.nbr = nbr
        self.tab = np.zeros(nbr)

    def echantillons(self):
        return self.nbr

    def moyenne(self):
        return np.mean(self.tab)

    def correlation(self):
        return np.corrcoef(self.tab, self.tab)[0, 1]

    def display(self):
        print(f"Nombre d'échantillons: {self.nbr}")
        print(f"Vecteur: {self.tab}")
        plt.plot(self.tab)
        plt.show()

class Aleatoire(Signal):
    def __init__(self, nbr, sigma, mean):
        Signal.__init__(self,nbr)
        self.sigma = sigma
        self.mean = mean
        self.init_alea()

    def init_alea(self):
        self.tab = np.random.randn(self.nbr) * self.sigma + self.mean

    def correlation(self):
        print("classe aléatoire")
        return np.power(self.tab, 2)

class Deterministe(Signal):
    def __init__(self, nbr, amplitude):
        Signal.__init__(self,nbr)
        self.amplitude = amplitude
        self.valeurs()

    def valeurs(self):
        for i in range(0,self.nbr):
            if i < self.nbr // 2:
                self.tab[i] = self.amplitude
            else:
                self.tab[i] = -self.amplitude

def Addition(objet1, objet2):
    result = Deterministe(objet1.nbr, 0)  
    result.tab = objet1.tab + objet2.tab
    return result





aleatoire = Aleatoire(20,1,0)

deterministe = Deterministe(20,2)

Nombre_échantillons_Aléatoire=aleatoire.echantillons()
print('le nombre de echantillon est :',Nombre_échantillons_Aléatoire)

moyenne_aléatoire=aleatoire.moyenne()
print('la moyenne aleatoire est:',moyenne_aléatoire)

corrélation_aléatoire=aleatoire.correlation()
print('corrélation aléatoire est:',corrélation_aléatoire)

aleatoire.display()



Nbre_échant_déterministe=deterministe.echantillons()
print('Nbre échant déterministe est :',Nbre_échant_déterministe)

moyenne_Déterministe=deterministe.moyenne()
print('moyenne Déterministe est :',moyenne_Déterministe)

Corr_Déterministe=deterministe.correlation()
print('Corr Déterministe est :',Corr_Déterministe)

deterministe.display()


resultat = Addition(aleatoire, deterministe)

signales=[aleatoire , deterministe ,resultat]
for signal in signales:
    signal.display()

```
